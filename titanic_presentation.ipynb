## 타이타닉호 데이터 분석 발표

```python
import pandas as pd  # 데이터 조작과 분석에 쓰는 라이브러리
import matplotlib.pyplot as plt  # 그래프 그릴 때 기본으로 쓰는 도구
import seaborn as sns  # 통계적 시각화를 쉽게 만들어주는 도구
```

```python
data = pd.read_csv('titanic1309.csv')  # 'titanic1309.csv' 파일을 읽어서 데이터프레임으로 저장
```

```python
print("기본 정보:")  # 데이터프레임의 기본 정보를 출력하기 위한 문구
print(data.info())   # 데이터의 컬럼별 타입, 결측치 개수, 메모리 사용량 등 요약 정보 출력
```

```python
print("기본 정보:")  # 데이터프레임의 기본 정보를 출력하기 위한 문구
print(data.info())   # 데이터의 컬럼별 타입, 결측치 개수, 메모리 사용량 등 요약 정보 출력
```

```python
print("데이터 샘플:")   # 데이터의 상위 5개 행을 보여주는 문구
print(data.head())      # 데이터의 처음 5개 행을 출력하여 내용 일부 확인
```

```python
survival_rate = data['Survived'].mean() * 100   # 'Survived' 컬럼의 평균 계산 → 생존 비율(%)
print(f"생존율: {survival_rate:.2f}%")   # 소수점 둘째 자리까지 생존율 출력
```

```python
survivors = data['Survived'].sum()      # 생존자 수: 'Survived' 컬럼에서 1인 값들의 합
non_survivors = len(data) - survivors   # 비생존자 수: 전체 데이터 수에서 생존자 수를 뺌
print(f"생존자 수: {survivors}")        # 생존자 수 출력
print(f"비생존자 수: {non_survivors}")  # 비생존자 수 출력
```

```python
survival_rate_by_gender = data.groupby('Sex')['Survived'].mean() * 100  # 'Sex'별로 그룹 지어 생존율(평균) 계산, 백분율로 변환

print("성별에 따른 생존율:")    # 성별 생존율 출력 안내 문구
print(survival_rate_by_gender)  # 성별별 생존율 출력
```

```python
plt.figure(figsize=(10, 6))  # 그래프 크기 지정 (가로 10, 세로 6)
sns.barplot(x='Sex', y='Survived', data=data)  # 성별에 따른 평균 생존율을 막대그래프로 표현
plt.title('성별에 따른 생존율')  # 그래프 제목 설정
plt.xlabel('성별')    # x축 라벨 설정
plt.ylabel('생존율')  # y축 라벨 설정
plt.show()  # 그래프 출력
```

```python
survival_rate_by_class = data.groupby('Pclass')['Survived'].mean() * 100  # 객실 등급(Pclass)별로 그룹 지어 생존율 계산, 백분율로 변환

print("객실 등급에 따른 생존율:")  # 객실 등급별 생존율 출력 안내 문구
print(survival_rate_by_class)      # 객실 등급별 생존율 출력
```

```python
plt.figure(figsize=(10, 6))  # 그래프 크기 지정 (가로 10, 세로 6)
sns.barplot(x='Pclass', y='Survived', data=data)  # 객실 등급별 평균 생존율을 막대그래프로 표현
plt.title('객실 등급에 따른 생존율')  # 그래프 제목 설정
plt.xlabel('객실 등급')  # x축 라벨 설정
plt.ylabel('생존율')     # y축 라벨 설정
plt.show()  # 그래프 출력
```

```python
data['Age'] = data['Age'].fillna(data['Age'].mean())  # 나이(Age) 결측값을 전체 평균나이로 대체

survival_rate_by_age = data.groupby('Age')['Survived'].mean() * 100  # 나이별로 그룹 지어 생존율 계산, 백분율로 변환

print("나이에 따른 생존율:")  # 나이별 생존율 출력 안내 문구
print(survival_rate_by_age)   # 나이별 생존율 출력
```

```python
plt.figure(figsize=(10, 6))  # 그래프 크기 지정 (가로 10, 세로 6)
sns.histplot(data['Age'], kde=True, bins=30)  # 나이 분포 히스토그램과 커널 밀도 추정(kde) 그리기, 구간 30개로 나눔
plt.title('나이 분포')  # 그래프 제목 설정
plt.xlabel('나이')  # x축 라벨 설정
plt.ylabel('빈도')  # y축 라벨 설정
plt.show()  # 그래프 출력
```

```python
plt.figure(figsize=(10, 6))  # 그래프 크기 지정 (가로 10, 세로 6)
sns.lineplot(x='Age', y='Survived', data=data)  # 나이별 생존율 추세를 선 그래프로 표현
plt.title('나이에 따른 생존율')  # 그래프 제목 설정
plt.xlabel('나이')   # x축 라벨 설정
plt.ylabel('생존율') # y축 라벨 설정
plt.show()  # 그래프 출력
```

```python
survival_rate_by_fare = data.groupby('Fare')['Survived'].mean() * 100  # 요금(Fare)별로 그룹 지어 생존율 계산, 백분율로 변환

print("요금에 따른 생존율:")  # 요금별 생존율 출력 안내 문구
print(survival_rate_by_fare)  # 요금별 생존율 출력
```

```python
survival_rate_by_embarked = data.groupby('Embarked')['Survived'].mean() * 100  # 승선 항구(Embarked)별로 그룹 지어 생존율 계산, 백분율로 변환

print("승선 항구에 따른 생존율:")  # 승선 항구별 생존율 출력 안내 문구
print(survival_rate_by_embarked)   # 승선 항구별 생존율 출력
```
